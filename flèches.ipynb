## <font color="red" > Tracer des flèches : <font>  


#### méthode 1 :  
On utilise plt.arrow. C'est à dire on trace une flèche. Ensuite , dans l'ordre on indique :
* l'abscisse du point de départ de la flèche x[i].
* l'ordonnée du  point de départ de la flèche y[i].
* l'abscisse du point d'arrivée de la flèche vx[i].
* l'ordonnée du point d'arrivée de la flèche vy[i].  

Le reste des instructions caractérise l'épaisseur, la couleur, la taille de la pointe,..).  
A la première éxécution, on voit que les flèches sont trop longues, donc on ajoute /10 pour ajuster l'échelle.

```python
import matplotlib.pyplot as plt

dt = 0.066
x = [0.003, 0.141, 0.275, 0.410, 0.554, 0.686, 0.820, 0.958, 1.089, 1.227, 1.359, 1.490, 1.599, 1.705, 1.801]
y = [0.746, 0.990, 1.175, 1.336, 1.432, 1.505, 1.528, 1.505, 1.454, 1.355, 1.207, 1.018, 0.797, 0.544, 0.266]


vx = []
vy = []
for i in range(len(x)-1):
    vx.append((x[i+1]-x[i])/dt)
    vy.append((y[i+1]-y[i])/dt)

plt.plot(x, y, ".")
for i in range (0,len(vx),2) :
     plt.arrow(x[i],y[i],vx[i]/10,vy[i]/10, shape='full', 
               lw=1,length_includes_head=True, rasterized=True, color="blue", head_width=.05,fc='blue')

    
plt.xlabel("x (m)")
plt.xlim(0, 2)
plt.ylabel("y (m)")
plt.ylim(0, 2)
plt.title("Trajectoire d'une balle")
plt.show()

```

#### méthode 2 :  
On utilise plt.quiver. C'est à dire on trace une flèche. Ensuite , dans l'ordre on indique :
* l'abscisse du point de départ de la flèche x[i].
* l'ordonnée du  point de départ de la flèche y[i].
* l'abscisse du point d'arrivée de la flèche vx[i].
* l'ordonnée du point d'arrivée de la flèche vy[i]. 

Documentation :  
https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.quiver.html

```python
import matplotlib.pyplot as plt

dt = 0.066
x = [0.003, 0.141, 0.275, 0.410, 0.554, 0.686, 0.820, 0.958, 1.089, 1.227, 1.359, 1.490, 1.599, 1.705, 1.801]
y = [0.746, 0.990, 1.175, 1.336, 1.432, 1.505, 1.528, 1.505, 1.454, 1.355, 1.207, 1.018, 0.797, 0.544, 0.266]


vx = []
vy = []
for i in range(len(x)-1):
    vx.append((x[i+1]-x[i])/dt)
    vy.append((y[i+1]-y[i])/dt)

plt.plot(x, y, ".")
for i in range (0,len(vx),2) :
    plt.quiver(x[i], y[i], vx[i], vy[i], angles='xy', scale_units='xy', scale=10, color='red', width=0.005)
    
plt.xlabel("x (m)")
plt.xlim(0, 2)
plt.ylabel("y (m)")
plt.ylim(0, 2)
plt.title("Trajectoire d'une balle")
plt.show()
```
