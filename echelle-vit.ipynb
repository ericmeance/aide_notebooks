```python
import matplotlib.pyplot as plt # bibliothèques servant aux graphes
dt =0.04
x = [-0.454077, -0.211902, 0.0151359, 0.232084, 0.443986, 0.671024, 0.887972, 1.10997, 1.33196, 1.54891, 1.7709, 1.99794, 2.21488, 2.43688, 2.65887]

y = [2.10389, 2.06353, 2.02316, 1.94748, 1.8718, 1.7709, 1.64981, 1.50854, 1.35214, 1.17555, 0.978787, 0.76184, 0.529756, 0.277491, -0.0050453]

plt.plot(x,y,'r+',ms=5.) # tracé des points,'r+' : points rouges sous forme de croix ms détermine la taille des marqueurs.

#on calcule les coordonnées des vitesses
vx=[]
vy=[]

for i in range (len(x)-1):
    vx.append((x[i+1]-x[i])/dt)
for i in range (len(y)-1):
    vy.append((y[i+1]-y[i])/dt)


#on trace les vecteurs vitesse, on introduit la variable vit
for i in range (0,len(vx),2) :
    vit=plt.quiver(x[i], y[i], vx[i], vy[i], angles='xy', scale_units='xy', scale=15, color='red', width=0.005)

# on indique l'échelle des vecteurs avec ( 0,-0.05) les coordonnées de l'étiquette et 5 la longueur de la flèche de l'échelle   
plt.quiverkey(vit,0,-0.05,5,label='echelle 5 m/s', coordinates='data' )# permet d'afficher l'échelle des vecteurs    

  
    
        
plt.xlabel("x (m)")
plt.xlim(min(x)-0.5, max(x)+0.5)
plt.ylabel("y (m)")
plt.ylim(min(y)-0.5, max(y)+0.5)
plt.title("Trajectoire d'une balle")
plt.show()
print(vit)
```
