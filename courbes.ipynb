## Tracer un graphe
Pour tracer un graphe, il faut utiliser des fonctions de tracé qui ne sont pas chargées par défaut. Il faut donc les appeler en  important  des **bibliothèques**  qui contiennent ces fonctions.

```python
import matplotlib.pyplot as plt # bibliothèques servant aux graphes
x= [0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y,'r+',ms=5.) # tracé des points,'r+' : points rouges sous forme de croix ms détermine la taille des marqueurs.
plt.xlim(min(x)-max(x)/100, max(x)+max(x)/10)  # on délimite les valeurs min et max de l'axe des abscisses
plt.ylim(min(y)-max(y)/100, max(y) +max(y)/10)
plt.xlabel("x en m") # on indique la grandeur sur l'axe des abscisses.
plt.ylabel("y en m") # on indique la grandeur sur l'axe des ordonnées.
plt.title( "trajectoire ") # on indique le titre.

plt.show()  # on affiche le graphe
```

On peut aussi tracer plusieurs courbes sur le même graphique et afficher le nom des grandeurs avec un label.

```python
x= [0,1,2,3,4]
y=[0,2,4,6,8]
x1 =x
y1=[0,3,6,9,12]
plt.plot(x,y,label= "courbe1",linestyle= 'dotted')
plt.plot(x1,y1,label= "courbe2" ,linestyle= 'dashdot')       # plot(abscisse,ordonnée,"legende de la courbe")


plt.xlabel("avancement en mol")      # xlabel("Nom de l'axe de l'axe des absicces")
plt.ylabel("quantités de matières en mol")              # ylabel("Nom de l'axe des ordonnées")
plt.legend(loc="upper right")           # Position de la legende de la courbe dans le graphique
plt.grid()                          # Affiche la grille du graphique
plt.show()
```
